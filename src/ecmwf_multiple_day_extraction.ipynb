```python
lat = 50.94
lon = 6.96

ds = xr.open_dataset('data/perturbed/enfo_pf_2024_05_13.nc')
vx = ds.interp(longitude=lon, latitude=lat)
mean_precipitation = vx.mean(dim='number')
std_precipitation = vx.std(dim='number')
```

```python
#index of times 
Mays18_index = ((mean_precipitation['time']>=pd.to_datetime('2024-05-18T00:00:00.000000000'))&
(mean_precipitation['time']<=pd.to_datetime('2024-05-19T00:00:00.000000000')))
#extract only the data of the day
vx = vx.sel(time= Mays18_index)
```
