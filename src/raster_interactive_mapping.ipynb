## Raster mapping  
To study the evolution of the forecast for the 18 of may of 2024, we use previusly produced rasters, and then mapped with leaflet and the folium library  

```python
import folium
from folium import plugins
from folium.raster_layers import ImageOverlay
import rasterio
from rasterio.plot import reshape_as_image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize, PowerNorm, LinearSegmentedColormap
import os
from functions import plots
```

```python
path='g:\\Mi unidad\\Universitat\\[SoSe24] Fundamentals of Earth System data processing\\ESDP_project\\Weather_forecast_case_stuty'
os.chdir(path)
```

```python
lat = 50.94
lon = 6.96
# Define the starting location and zoom level for the map
m = folium.Map(location=[lat, lon], zoom_start=1)

# Create colormap
colormap = plots.create_colormap()

raster1 = 'data/raster/era5_new_units.tif'
raster2 = 'data/raster/enfo_pf_2024_05_17.tif'
raster3 = 'data/raster/enfo_pf_2024_05_16.tif'
raster4 = 'data/raster/enfo_pf_2024_05_15.tif'
raster5 = 'data/raster/enfo_pf_2024_05_14.tif'
raster6 = 'data/raster/enfo_pf_2024_05_13.tif'

plots.add_raster_to_map(m, raster1, 'Era5 reanalysis data 18/05', colormap)
plots.add_raster_to_map(m, raster2, '17/05 forecast for 18/05', colormap)
plots.add_raster_to_map(m, raster3, '16/05 forecast for 18/05', colormap)
plots.add_raster_to_map(m, raster4, '15/05 forecast for 18/05', colormap)
plots.add_raster_to_map(m, raster5, '14/05 forecast for 18/05', colormap)
plots.add_raster_to_map(m, raster6, '13/05 forecast for 18/05', colormap)


# Add layer control
folium.LayerControl(collapsed=True).add_to(m)

# Save the map as an HTML file
m.save('interactive_map_red.html')
m
```
